```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import csv

# File to Load (Remember to Change These)
food_data_to_load = "Resources/food_access_and_population_2015.csv"
unemployment_data_to_load = "Resources/unemployment_2007-2017.csv"
county_data_to_load = "Resources/cities_and_counties.csv"
murder1516_data_to_load = "Resources/murder_rates_2015-16.csv"

# Read School and Student Data File and store into Pandas Data Frames
food_data = pd.read_csv(food_data_to_load)
unemployment_data = pd.read_csv(unemployment_data_to_load)
county_data = pd.read_csv(county_data_to_load)
murder1516_data = pd.read_csv(murder1516_data_to_load)
```

```python
food_state = food_data['State']
food_county = food_data['County']
food_pop = food_data['Population']
food_people_half = food_data['People.1/2 Mile']
food_people_one = food_data['People.1 Mile']
food_people_ten = food_data['People.10 Miles']
food_people_twenty = food_data['People.20 Miles']

fooddata_clean = pd.DataFrame(
    {"State": food_state,"County": food_county, "Population": food_pop, 
      "People 1/2 Mile and More": food_people_half, "People 1 Mile and More": food_people_one, 
      "People 10 Miles and More": food_people_ten, "People 20 Miles and More": food_people_twenty})

fooddata_clean.head()
```

```python
unemployment_clean.to_csv('Resources/Clean_Data/unemployment_data.csv')
```
