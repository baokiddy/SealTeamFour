```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
# importing dataframes from 'project_dataclean'
city_data = pd.read_csv("Resources/Clean_Data/CityData/city_data_rent_murders_population.csv")
county_data = pd.read_csv("Resources/Clean_Data/CountyData/county_data_food_unemployment.csv")
```

```python
county_data.head()
```

```python
# rate of unemployment vs food scarcity 1/2 mile (county)
plt.figure(figsize=(20, 20))
plt.scatter(county_data["Rate of Unemployment"], county_data["People 1/2 Mile and More"],
            marker="o", facecolors="red", edgecolors="black")

# Create a title, x label, and y label for our chart
plt.title("Unemployment Rate (%) v. People 1/2 Mile and More")
plt.ylabel("People 1/2 Mile and More")
plt.xlabel("Unemplyment Rate (%)")
plt.xticks(rotation=90)

#plt.xlim(-80, 100)
#plt.ylim(-100, 150)
plt.grid()

# # Save the image of the chart and print to screen
plt.savefig('CuontyPopUnmpRate.png')
plt.show()
```

```python
# rate of unemployment vs food scarcity 1/2 mile (county)
# plt.figure(figsize=(20, 20))

pop_data = [county_data["People 1/2 Mile and More"], county_data["People 1 Mile and More"],
            county_data["People 10 Miles and More"],
            county_data["People 20 Miles and More"]]
plt.scatter(pop_data, county_data["Rate of Unemployment"],
            marker="o", facecolors="red", edgecolors="black")

# Create a title, x label, and y label for our chart
plt.title("Unemployment Rate (%) v. Miles")
plt.xlabel("Miles")
plt.ylabel("Unemplyment Rate (%)")
# plt.xticks(rotation=90)

#plt.xlim(-80, 100)
#plt.ylim(-100, 150)
plt.grid()

# # Save the image of the chart and print to screen
plt.savefig('CountyPopUnmpRate.png')
plt.show()
```

```python
# Scatter plot to showcase the relationship between County Population in USA & THE Rate of Employment in that County
# (marker="o", facecolors="red", edgecolors="black", s=scoop_price)

plt.figure(figsize=(20, 20))
plt.scatter(county_data["Population"], county_data["Rate of Unemployment"], marker="o", facecolors="red", edgecolors="black")

# Create a title, x label, and y label for our chart
plt.title("Unemployment Rate (%)v. USA County Population")
plt.xlabel("County Population")
plt.ylabel("Unemplyment Rate (%)")
plt.xticks(rotation=90)

#plt.xlim(-80, 100)
#plt.ylim(-100, 150)
plt.grid()

# # Save the image of the chart and print to screen
plt.savefig('CuontyPopUnmpRate.png')
plt.show()
```
